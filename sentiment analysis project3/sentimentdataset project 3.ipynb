```python
plt.figure(figsize=(8,6))
data.head(30)['Sentiment'].value_counts().plot(kind='bar')
plt.title('Distribution of sentiment labels')
plt.xlabel('sentiment')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
```

In the above graph it shows high positive sentiment in distribution of sentiment label.

```python
data['word_count'] = data['Text'].apply(lambda x: len(str(x).split()))
data['word_count'].head(20).plot(kind='hist', bins=20, color='lightgreen', figsize=(6,4))
plt.title('Distribution of Word Count in Posts')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.show()`
```

The histogram of word count in posts reveals the frequency distribution, indicating whether most posts are short, medium, or long in length, which can inform the overall engagement and style of the content

```python
platform_user_count = data.groupby('Platform')['User'].nunique().sort_values()
plt.figure(figsize=(10, 6))
platform_user_count.plot(kind='bar', color='skyblue')
plt.title('Number of Unique Users by Platform')
plt.xlabel('Platform')
plt.ylabel('Number of Unique Users')
plt.xticks(rotation=45)
plt.show()
```

highest number of unique user is on instagram platform.

```python
data['Sentiment'].head(20).value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), colors=['lightgreen', 'salmon', 'gold'])
plt.title('Proportion of Sentiments')
plt.ylabel('')
plt.show()
```

```python
data['Hashtags'].head(10).value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), colors=['lightgreen', 'salmon', 'gold'])
plt.title('Proportion of Hashtags')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data.head(30)['Year'],data.head(30)['Likes'])
plt.title('years per likes')
plt.xlabel('year')
plt.ylabel('likes')
plt.show()
```

In the above graph 2002 to 2025 having more likes.

```python
plt.figure(figsize=(10,7))
platform_country_count = data.head(30).groupby(['Country', 'Platform']).size().unstack()
platform_country_count.plot(kind='bar', stacked=True, figsize=(10,6), color=['purple', 'orange', 'green', 'blue'])
plt.title('Platform Usage by Country')
plt.xlabel('Country')
plt.ylabel('Platform Count')
plt.xticks(rotation=45)
plt.legend(title='Platform', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Sentiment vs. Length of Posts:
# Calculate the length of each post (word count)
data['text_length'] = data['Text'].apply(lambda x: len(str(x).split()))
# Create a boxplot to compare text length across different sentiments
data.head(10).boxplot(column='text_length', by='Sentiment', grid=False, figsize=(10,6))
# Set plot labels and title
plt.title('Sentiment vs. Length of Posts')
plt.suptitle('')  # Remove default title
plt.xlabel('Sentiment')
plt.ylabel('Text Length (Word Count)')
plt.show()
```
